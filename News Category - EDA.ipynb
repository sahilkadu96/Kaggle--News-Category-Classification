```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from wordcloud import WordCloud, STOPWORDS
```

```python
df = pd.read_json('News_Category_Dataset_v3.json', lines = True)
df
```

```python
df.info()
```

```python
df['category'].unique()
```

```python
category_label = df.category.value_counts()[:5]
plt.figure(figsize=(10, 5))
sns.barplot(category_label.index, category_label)
plt.xlabel('category')
plt.ylabel('count');
plt.title('Top 5 categories')
```

```python
stopwords = set(STOPWORDS)
```

```python
def generateWordCloud(dataframe, categ):
    dataframe_cat = dataframe[dataframe['category'] == categ]
    words = dataframe_cat['short_description'].tolist()
    combined_words = ' '.join(words)
    plt.figure(figsize = (10, 10))
    wordcloud = WordCloud(background_color = 'white', stopwords = stopwords).generate(combined_words)
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
    
```

```python
generateWordCloud(df, 'POLITICS')
```

```python
generateWordCloud(df, 'WELLNESS')
```

```python
generateWordCloud(df, 'ENTERTAINMENT')
```

```python
generateWordCloud(df, 'TRAVEL')
```
